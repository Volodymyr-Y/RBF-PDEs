```julia
include("RBFunctions.jl")
using GLMakie
using CairoMakie
```

```julia
N_coll = 60
N_eval = 50
coll_points = generate_2D_Halton_points(N_coll)
eval_points= zeros(Float64,(2,N_eval^2))
a= LinRange(0.0,1.0,N_eval)
eval_points[1,:] = vec(ones(N_eval) * a')
eval_points[2,:] = vec( a* ones(N_eval)')
1
```

```julia
function error(RBF,shape_par,eval_p,coll_p,f_true) # calculate error with given evaluation points and collocation points 
    A = pairwise(Euclidean(), coll_p, coll_p, dims=2)
    A = RBF.(A,shape_par)
    coeff = inv(A) * f_true.(coll_p[1,:],coll_p[2,:]) # find coefficients
    # reconstruct function at evaluation points 
    B = pairwise(Euclidean(), eval_p, coll_p,  dims=2)
    B = RBF.(B,shape_par)
    f_reconstructed = B*coeff
    error_vec = abs.(f_reconstructed-f_true.(eval_p[1,:],eval_p[2,:]))
    error = sqrt(((norm(error_vec))^2)/length(error_vec))
    #error = maximum(error_vec)
    #println(size(f_reconstructed))
    return error
end 

function err_Halton(N,shape_par,RBF,f_true) # error as a function of shape parameter and N 
    N_eval = 200
    coll_points = generate_2D_Halton_points(N)
    eval_points= zeros(Float64,(2,N_eval^2))
    a= LinRange(0.0,1.0,N_eval)
    eval_points[1,:] = vec(ones(N_eval) * a')
    eval_points[2,:] = vec( a* ones(N_eval)')
    return error(RBF,shape_par,eval_points,coll_points,f_true)
end


error(gaussian,2,eval_points,coll_points,frankes_func)

err_Halton(60,2,gaussian,frankes_func)
```

```julia
ϵ = LinRange(0.5,20,40)
N_list = [10,20,50,100]
f = Figure(resolution = (600, 300))
ax = Axis(f[1, 1],yscale=log10)
ax.ylabel = "RMS error"
ax.xlabel = "shape parameter"
for N in N_list
    lines!(ax,ϵ, err_Halton.(N,ϵ,gaussian,frankes_func),label = "N="*string(N))
end
axislegend()
f
```

### Cross Validation 


```julia
function cross_validation_error(RBF,shape_par,coll_p,f_true)
    A = pairwise(Euclidean(), coll_p, coll_p, dims=2)
    A = RBF.(A,shape_par)
    inv_A = inv(A)
    coeff = inv_A * f_true.(coll_p[1,:],coll_p[2,:]) # find coefficients
    error_vec = coeff/diag(inv_A)
    #error = maximum(error_vec)
    error = sqrt((norm(error_vec)^2)/length(error_vec))
    return error
end
function cv_err_Halton(N,shape_par,RBF,f_true)
    coll_points = generate_2D_Halton_points(N)
    return cross_validation_error(RBF,shape_par,coll_points,f_true)
end
```

```julia
ϵ = LinRange(0.5,20,40)
N_list = [10,20,50,100]
f2 = Figure(resolution = (600, 300))
ax2 = Axis(f2[1, 1],yscale=log10)
ax2.ylabel = "RMS error"
ax2.xlabel = "shape parameter"
for N in N_list
    lines!(ax2,ϵ, cv_err_Halton.(N,ϵ,gaussian,frankes_func),label = "N="*string(N))
end
axislegend()
f2
```
